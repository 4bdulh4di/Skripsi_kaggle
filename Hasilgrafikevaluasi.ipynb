```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, confusion_matrix

# Contoh nilai sebenarnya dan prediksi model
y_true = [0, 1, 0, 1, 0, 1, 0, 1]  # Label sebenarnya
y_pred = [0, 1, 0, 0, 1, 1, 0, 1]  # Prediksi model

# Menghitung metrik evaluasi
accuracy = accuracy_score(y_true, y_pred)
precision = precision_score(y_true, y_pred, average='macro')
recall = recall_score(y_true, y_pred, average='macro')
f1 = f1_score(y_true, y_pred, average='macro')

# Menghitung confusion matrix
cm = confusion_matrix(y_true, y_pred)

# Membuat tabel hasil evaluasi
metrics = ['Akurasi', 'Presisi', 'Recall', 'F1-score']
values = [accuracy, precision, recall, f1]

eval_df = pd.DataFrame({'Metrik': metrics, 'Nilai': values})
print("Tabel Hasil Evaluasi:")
print(eval_df.to_string(index=False))

# Membuat grafik bar untuk metrik evaluasi
plt.figure(figsize=(10, 6))
sns.barplot(x='Metrik', y='Nilai', data=eval_df)
plt.title('Metrik Evaluasi Model KNN')
plt.ylim(0, 1)
for i, v in enumerate(values):
    plt.text(i, v, f'{v:.4f}', ha='center', va='bottom')
plt.show()

# Membuat heatmap untuk confusion matrix
plt.figure(figsize=(8, 6))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues')
plt.title('Confusion Matrix')
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.show()
```
